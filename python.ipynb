Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a maximum score of 800 points. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance:

## Which schools produce the highest math scores?
* Specifically, which schools have an average math SAT score of at least 80%?
* Save the results as a pandas DataFrame called `best_math_schools`.

## Who are the top 10 schools based on average results across reading, math, and writing?
* Save the results as a pandas DataFrame called `top_10_schools`.

## Which NYC borough has the largest standard deviation for SAT results?
* Save the results as a pandas DataFrame called `largest_std_dev`.

## Completing the project

* Create a dictionary called `schools_analysis` containing the following key-value pairs:
	- `"best_math_schools"`: `best_math_schools` DataFrame.
	- `"top_10_schools"`: `top_10_schools` DataFrame.
    - `"largest_SAT_std_dev"`: `largest_std_dev` DataFrame.

```python
# Start coding here... 
import pandas as pd
```

```python
#load dataset
data = pd.read_csv("schools.csv")
data.head()
```

Find the "school_name" and "average_math" score for all schools where the results are at least 80% of the maximum, sorting by "average_math" in descending order, and storing as a pandas DataFrame called best_math_schools.

```python
best_math_school = data[data["average_math"] >= 640][["school_name", "average_math"]].sort_values("average_math", ascending=False)
```

Identify the top 10 performing schools based on scores across the three SAT sections, storing as a pandas DataFrame called top_10_schools containing the school name and a column named "average_SAT".

```python
data["average_SAT"] = data["average_math"] + data["average_reading"] + data["average_writing"]

top_10_schools = data.sort_values("average_SAT", ascending=False)[["school_name", "average_SAT"]].head(10)
                                                                            
```

Locate the NYC borough with the largest standard deviation for "average_SAT", storing as a DataFrame called largest_std_dev containing the borough name as the index, and columns called "count" for the number of schools, "mean" for the mean of "average_SAT", and "std" for the standard deviation of "average_SAT".

```python
boroughs = data.groupby("borough")["average_SAT"].agg(["count", "mean", "std"]).round(0).sort_values("mean", ascending=False)
largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]
```

```python
# Share your results in this format
schools_analysis = {"best_math_schools": best_math_school,
                    "top_10_schools": top_10_schools,
                    "largest_SAT_std_dev": largest_std_dev}

# Call the answer!
print(schools_analysis)
```
